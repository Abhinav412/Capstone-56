```python
cols_to_drop = ['Year End', 'Company Type', 'PAT', 'PBIDT', 'PBIT', 'EPS (Annualised) (Adjusted) (Unit Curr)', 'ROCE (%).1'] # Replace with your column names
df = df.drop(columns=cols_to_drop)
df.columns
```

## Data cleaning

Clean the data by handling missing values and zero values.


```python
# Identify columns with missing values
missing_values_count = df.isnull().sum()
columns_with_missing_values = missing_values_count[missing_values_count > 0].index.tolist()

# Handle missing values for numerical columns
for column in columns_with_missing_values:
  if pd.api.types.is_numeric_dtype(df[column]):
    df[column] = df[column].fillna(df[column].median())

# Handle missing values for categorical columns (replace with most frequent)
for column in columns_with_missing_values:
  if not pd.api.types.is_numeric_dtype(df[column]):
    df[column] = df[column].fillna(df[column].mode()[0])
print("\nNumber of missing values in each column:\n", df.isnull().sum())
```

```python
# Count zero values in each numerical column
zero_counts = df.select_dtypes(include='number').apply(lambda x: (x == 0).sum())
print("\nNumber of zero values in each numerical column:\n", zero_counts)
```

##Data Visualization



```python
# prompt: give a heatmap for all the numerical features

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame as defined in the previous code.

# Select only numerical features for the heatmap
numerical_features = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Create the heatmap
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# prompt: since gross sales net sales and value of output are highly correlated do feature extraction

# Create a new feature representing the average of 'Gross Sales', 'Net Sales', and 'Value of Output'
df['Average_Sales_Output'] = (df['Gross Sales'] + df['Net Sales'] + df['Value Of Output']) / 3

# Drop the original highly correlated features
df = df.drop(['Gross Sales', 'Net Sales', 'Value Of Output'], axis=1)

# Display the updated DataFrame
display(df.head())
```

```python

df['Year'] = df['Year'].astype(int)
# Verify the data type change
print("\nData types of columns:\n", df.dtypes)
```

```python
# prompt: I want to know the number of companies that are there in each year give me a bargraph

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and 'Year' is the column representing the year
company_counts_by_year = df.groupby('Year')['NSE Symbol'].nunique()

plt.figure(figsize=(12, 6))
sns.barplot(x=company_counts_by_year.index, y=company_counts_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Companies')
plt.title('Number of Unique Companies per Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
FINANCIAL_FEATURES = [
    'Debt-Equity Ratio', 'Current Ratio', 'Net Working Capital ( Incl. Def. Tax)',
    'EPS (annualised) (Unit Curr)', 'ROCE (%)', 'ROE (X)', 'Interest Cover Ratio',
    'Market Cap/Sales', 'PBIDTM (%)', 'APATM (%)', 'Price Earning (P/E)'
]
corr_matrix = df[FINANCIAL_FEATURES].corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f",
            mask=np.triu(np.ones_like(corr_matrix)))
plt.title('Financial Feature Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
sector_counts = df['Sector'].value_counts()

plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
sector_counts.plot(kind='bar')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.title('Distribution of Companies by Sector')
plt.xticks(rotation=90, ha='right', fontsize=10)  # Rotate x-axis labels by 90 degrees
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('/content/processed_data.csv')

# Get a list of unique companies
unique_companies = df['Company Name'].unique()
filtered_companies = []

for company in unique_companies:
  company_data = df[df['Company Name'] == company]  # Replace 'Company Name' with the actual column name

  # Check if financial data is available for all years from 2010 to 2024
  years_available = company_data['Year'].unique()  # Replace 'Financial Year' with the actual column name
  if all(year in years_available for year in range(2010, 2025)):
    filtered_companies.extend(company_data.index.tolist())  # If all years are available, append the rows to filtered list

# Create a new DataFrame with filtered companies
df_filtered = df.loc[filtered_companies]

display(df_filtered)
```
